The list of unique partners is critical for constructing the graph. The free-text field in the survey makes it easy to introduce lots of different variations into the partner list.

This script hunts for all the partners listed - combining the of the organisation completing the survey, with the list of the partners they collaborate with. It expects a file called `input/test-lime-data.csv`. Check the `output/unique_partners.csv` file after it has run. (It needs an existing `output` folder to exist before it runs).

This list can be used as a diagnostic tool to hunt for some obvious renaming needed.

```python
import pandas as pd
df = pd.read_csv('input/test-lime-data.csv')
```

```python
all_headers = pd.DataFrame(df.columns.tolist())
all_headers.to_csv('output/headers.csv')
```

```python
partner_name_indices = [0] # Start with the org's own name
partner_name_indices.extend(list(range(42, 52))) # Add column indices for partners
partner_name_indices
partner_names = df.iloc[:, partner_name_indices]
partner_names
```

```python
unique_partner_names = partner_names.melt().dropna().sort_values('value').reset_index().loc[:, ['value']]
unique_partner_names.index.name = 'id'
unique_partner_names.columns = ['partner_name']
unique_partner_names
```

```python
unique_partner_names.to_csv('output/unique_partners.csv')
```
